```python
def x_data():
    mu_l, sigma_l = [0, 0.5, 0.6, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35], [0.1, 0.1, 0.1, 0.01, 0.01, 0.01,  0.01, 0.01, 0.01]

    X = np.empty((0, 2), dtype=np.float)
    for mu, sigma in zip(mu_l, sigma_l):
        x1 = np.random.normal(mu, sigma, 1000).reshape(-1, 2)
        X = np.concatenate((X, x1), axis=0)

    for mu, sigma in zip(mu_l[:3], sigma_l[:3]):
        x1 = np.random.normal(mu, sigma, 10000).reshape(-1, 2)
        X = np.concatenate((X, x1), axis=0)
    return X
```

```python
kde = KernelDensityExp(kernel='gaussian', bandwidth=0.5).fit(x_t)
```

```python
x_exp=[[0.0, 0.0]]
x_exp = np.array(x_exp)
# mean and standard deviation
X = x_t

plt.scatter(X[:, 0], X[:, 1], c='gray', s=10)
N = 200
x = np.linspace(-25.0, 25.0, N)
y = np.linspace(-25.0, 25.0, N)
X, Y = np.meshgrid(x, y)
xy = np.vstack([X.ravel(), Y.ravel()]).T
Z = kde.score_samples(xy)
Z = np.exp(Z.reshape(N, N))
levels = np.linspace(0, Z.max()/3, 15)
print(Z.max())
plt.contour(X, Y, Z, levels=levels, colors='black')
# plt.contour(X, Y, Z, colors='black')
plt.grid(True, which='both')
```

```python
def gausss(x, mu, sigma):
    d = np.sqrt(np.sum((x-mu)**2, axis=1))
    normalize = (sigma*np.sqrt(2.0*np.pi))**-1
    return normalize*np.exp(-0.5*d**2/sigma**2)
```

```python
x = np.linspace(-5, 5, 1000).reshape(-1, 1)
y = gausss(x, 0, sigma=0.5).reshape(-1, 1)
```
